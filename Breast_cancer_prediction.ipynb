```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
dataFrame = pd.DataFrame(breast_cancer_dataset.data,columns = breast_cancer_dataset.feature_names)
```

```python
dataFrame['label'] = breast_cancer_dataset.target
```

```python
X = dataFrame.drop(columns = 'label',axis=1)
Y = dataFrame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

standardise the data

```python
 from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

Building the neural network

We will import libraries for making the neurla network that are tensorflow and keras

```python
import tensorflow as tf
tf.random.set_seed(3)
# it gives us same accuracy score for each time it is run
from tensorflow import keras
```

```python
from keras.api._v2.keras import activations
#setting up the layers of neural network
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20,activation='relu'),
                          keras.layers.Dense(30,activation='sigmoid'),
                          keras.layers.Dense(2,activation = 'sigmoid')
                          # 2 represent the number of possible classes of outputs that are possible in output layer
                          ])
```

```python
# compiling the neural network

model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              # we have two types of loss functions
              metrics = ['accuracy'])
```

training the neural network

```python
# validation_Split is like train_test_Split only and used for the testing
# epochs are the number of times the model has to go through the data
history = model.fit(X_train_std,Y_train,validation_split=0.1,epochs = 10)
```

Visualising the accuracy and the loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
loss,accuracy = model.evaluate(X_test_std,Y_test)
print(accuracy)
```
